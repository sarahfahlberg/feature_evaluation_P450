```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from regression_tools import *

#Get one-hot-encoding data
AA_seqs, thermo_data = pickle.load(open('../P450_Chimeras/save.p', 'rb'))
y_data = [float(y) for _, y, _ in thermo_data]
parents = AA_seqs[:3]

#Encode sequences
AA_seqs_encoded, pos_coding = oneHotEncodeOptimized(AA_seqs, parents, standardize = True)
AA_seqs_decoded = decodeOneHotEncodeOptimized(AA_seqs_encoded, pos_coding, parents, standardize = True)

checkEncoding(AA_seqs, parents, standardize = True)

#Get embeddings
seq_file = open('embeddings_reproduction/outputs/T50_n_embeddings/X_original_3_7_160000.pkl',"rb")
p_embeddings, _ = pickle.load(seq_file)

#Sequences 242-beyond are not included in the original dataset, so takes 0-241
embeddings_raw = p_embeddings.to_numpy()[:242]
blocks = p_embeddings.index.values[:242]

#Matches embeddings to function
embeddings = []
blocks_y_data = [[block, float(y)] for block, y, _ in thermo_data]
blocks_ind_0 = [[str(int(block)-11111111), y] for block, y in blocks_y_data]
for [block_y, y], block_e, embed in zip(blocks_ind_0, blocks, embeddings_raw):
    assert block_y == str(int(block_e)) #Exception thrown if data does not match up
    embeddings.append(embed.tolist())
```

```python
# Leave-One-Out CV function check 
y_hats, r, mserr, mae = leaveOneOutCV_Ridge_Std(embeddings, y_data)

import matplotlib.pyplot as plt
plt.plot(y_data,y_hats,'o')
plt.plot([40,65],[40,65], linestyle = '--', color = 'k')
plt.text(39, 64, 'r = ' + str('%0.3f' % r) + '\nMSE = ' + str('%0.3f' % mserr) + '\nMAE = ' + str('%0.3f' % mae), 
         verticalalignment='center')
plt.ylabel('Predicted T50')
plt.xlabel('Actual T50')
plt.show()
```

```python
#Tests many lambdas over logspace - only run this to recalculate lambda values, otherwise
#it will pull lambdas calculated from python file previously from a pickle file

'''
#Lambda for one-hot encoding and for embeddding
lambdas, mserrs_enc, maes_enc = selectLambda(AA_seqs_encoded, y_data)
lambdas, mserrs_emb, maes_emb = selectLambda(embeddings, y_data)
with open('lambdas.pkl', 'wb') as file:
    pickle.dump([lambdas, mserrs_enc, maes_enc, mserrs_emb, maes_emb], file)
'''

#Loads lambdas from a pickle file
with open('lambdas.pkl', 'rb') as f:
     lambdas, mserrs_enc, maes_enc, mserrs_emb, maes_emb = pickle.load(f)

log_lambdas = []
for l in lambdas:
    log_lambdas.append(np.log10(l))

#Plots MAE versus log(lambda) which can be used to find optimal lambda
plt.plot(log_lambdas, mserrs_enc)
plt.plot(log_lambdas, mserrs_emb)
plt.xlabel('log(lamdba)')
plt.ylabel('MSE')
plt.legend(['One-hot encoding', 'Learned embeddings'])
#plt.axis([-10,2,5.5,7])

# Selects a standard lambda and prints the lambda
std_l_enc = lambdas[mserrs_enc.index(min(mserrs_enc))]
std_l_emb = lambdas[mserrs_emb.index(min(mserrs_emb))]

print([std_l_enc, std_l_emb])

plt.plot(np.log10(std_l_enc), mserrs_enc[mserrs_enc.index(min(mserrs_enc))], 'o', color = 'cornflowerblue')
plt.plot(np.log10(std_l_emb), mserrs_emb[mserrs_emb.index(min(mserrs_emb))], 'o', color = 'orange')
plt.show()
```
